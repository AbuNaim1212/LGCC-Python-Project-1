```python
import numpy as np
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.count()
```

```python
# 1st School 'Orchard Collegiate Academy'
df1=df[(df['School Name'] == 'Orchard Collegiate Academy') & (df['Regents Exam'] == 'English')]
print(df1)
```

```python
df2=df1.groupby(['Percent Scoring CR'])['Percent Scoring CR'].count().sort_values(ascending=False)
print(df2)
```

```python
df2.plot(kind="bar")
```

```python
# 2nd School 'Manhattan / Hunter Science High School'
df3=df[(df['School Name'] == 'Manhattan / Hunter Science High School') & (df['Regents Exam'] == 'English')]
print(df3)
```

```python
df4=df3.groupby(['Percent Scoring CR'])['Percent Scoring CR'].count().sort_values(ascending=False)
print(df4)
```

```python
df4.plot(kind='bar')
```

```python
#Findings:
1) In a sense of 's(5 or fewer students)' Percent Scoring CR is same for both school based on English Test.
2) But in a sense of whole comparision of 'Percent Scoring CR' Hunter Science High School has higher percentage(Ex: 100,97.1,96.7 etc)
   than Orchard Collegiate Academy percentage(Ex: 38.9,27.8,27.7).
* So on the above comparative study we can say 'Hunter Science High School' has superior performance than 'Orchard Collegiate Academy'
  in College Ready segment based on English Test.
```
